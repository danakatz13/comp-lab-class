```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
# Load the trajectory
traj = md.load_trr('../Inputs/md_adp.trr', top='../Inputs/md_adp.gro')
```

```python
#Compute the phi and psi angles of the trajectory
phi, psi = md.compute_phi(traj), md.compute_psi(traj)
```

```python
#plot phi and psi
plt.plot(phi[1][:,0], label='Phi (Degrees)')
plt.plot(psi[1][:,0], label='Psi (Degrees)')
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.show()
```

```python
#compute 2d histogram of phi and psi
H, xedges, yedges = np.histogram2d(phi[1][:,0], psi[1][:,0], bins=100)

#plot 2d histogram
plt.imshow(H, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.colorbar()
plt.show()
```

```python
#plot -log of histogram
plt.imshow(-np.log(H), interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
#add axis and title
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('2D Histogram of Phi and Psi')
#add colorbr
plt.colorbar()
```

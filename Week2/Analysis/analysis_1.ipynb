```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
data_np = np.loadtxt('../Data/1hz3_T310.run.25000000.energy.xvg', skiprows=29)
data = pd.DataFrame(data_np, columns=['Time', 'Potential Energy', 'Kinetic Energy', 
'Total Energy','Temperature','Volume','Pressure'])
```

```python
data
```

```python
# Make 6 separate plots for each quantity vs time, with time in ns
fig, axes = plt.subplots(2,3, figsize=(23,23))
sns.lineplot(data=data, x="Time",y="Potential Energy", ax=axes[0,0])
sns.lineplot(data=data, x="Time",y="Kinetic Energy", ax=axes[0,1])
sns.lineplot(data=data, x="Time",y="Total Energy", ax=axes[0,2])
sns.lineplot(data=data, x="Time",y="Temperature", ax=axes[1,0])
sns.lineplot(data=data, x="Time",y="Pressure", ax=axes[1,1])
sns.lineplot(data=data, x="Time",y="Volume", ax=axes[1,2])
#add axis titles
axes[0,0].xaxis.set_label_text("Time (ns)")
axes[0,0].yaxis.set_label_text("Potential Energy (kJ/mol)")
axes[0,1].xaxis.set_label_text("Time (ns)")
axes[0,1].yaxis.set_label_text("Kinetic Energy (kJ/mol)")
axes[0,2].xaxis.set_label_text("Time (ns)")
axes[0,2].yaxis.set_label_text("Total Energy (kJ/mol)")
axes[1,0].xaxis.set_label_text("Time (ns)")
axes[1,0].yaxis.set_label_text("Temperature (K)")
axes[1,1].xaxis.set_label_text("Time (ns)")
axes[1,1].yaxis.set_label_text("Pressure (Pa)")
axes[1,2].xaxis.set_label_text("Time (ns)")
axes[1,2].yaxis.set_label_text("Volume" r'$(nm^3)$')
```

```python
#Make a plot that has kinetic energy and temperature on the same plot, where the left axis is temperature and the right axis is kinetic energy, using different colors and different axis labels
fig, ax1 = plt.subplots(figsize=(10,10))
ax2 = ax1.twinx()
sns.lineplot(data=data, x="Time",y="Temperature", ax=ax1, color="red")
sns.lineplot(data=data, x="Time",y="Kinetic Energy", ax=ax2, color="blue")
ax1.xaxis.set_label_text("Time (ns)")
ax1.yaxis.set_label_text("Temperature (K)")
ax2.yaxis.set_label_text("Kinetic Energy (kJ/mol)")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mdtraj as md
```

```python
data = np.loadtxt('../sim_A/metadynamics_2d/fes.dat', comments=['#'])
data_B = np.loadtxt('../sim_B/metadynamics_2d/fes.dat', comments=['#'])
```

```python
#reshape the data
phi_new_A = data[:33,0]
psi_new_A = data[::33,1]
fes_A = data[:,2]
fes_A_new = fes_A.reshape(18,33)


phi_new_B = data_B[:33,0]
psi_new_B = data_B[::33,1]
fes_B = data_B[:,2]
fes_B_new = fes_B.reshape(18,33)
```

```python
print(fes_A.shape)
```

```python
#plot contour of A and B using subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.contourf(phi_new_A, psi_new_A, fes_A_new, 100, cmap='viridis')
ax1.set_xlabel(r'$\phi$')
ax1.set_ylabel(r'$\psi$')
ax1.set_title('A')
ax2.contour(phi_new_B, psi_new_B, fes_B_new, 100, cmap='viridis')
ax2.set_xlabel(r'$\phi$')
ax2.set_ylabel(r'$\psi$')
ax2.set_title('B')
plt.show()
```

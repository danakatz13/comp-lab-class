```python
import matplotlib.pyplot as plt
import numpy as np
import mdtraj as md
```

```python
data_A = np.loadtxt("../sim_A/COLVAR")
data_B = np.loadtxt("../sim_B/COLVAR")
```

```python
#import trajectories
traj_A = md.load("../sim_A/traj_comp.xtc", top="../Inputs/dialaA.pdb")
traj_B = md.load("../sim_B/traj_comp.xtc", top="../Inputs/dialaB.pdb")
```

```python
#compute phi and psi from trajectory for A and B
phi_A = md.compute_phi(traj_A)
psi_A = md.compute_psi(traj_A)
phi_B = md.compute_phi(traj_B)
psi_B = md.compute_psi(traj_B)
```

```python
#plot phi and psi for A and B using data
plt.plot(data_A[:,1], data_A[:,2], 'o', label="A")
plt.plot(data_B[:,1], data_B[:,2], 'o', label="B")
plt.xlabel("Phi (radians)")
plt.ylabel("Psi (radians")
plt.legend()
plt.show()
```

```python
#plot phi vs psi for A and B on same plot
plt.plot(phi_A[1], psi_A[1], 'o', label='A')
plt.plot(phi_B[1], psi_B[1], 'o', label='B')
plt.xlabel('Phi (radians')
plt.ylabel('Psi (radians')
plt.legend()
plt.show()
```

```python
#plot md traj data
plt.plot(phi_A[1], psi_A[1], 'o', label='A')
plt.plot(phi_B[1], psi_B[1], 'o', label='B')
plt.xlabel('Phi (radians')
plt.ylabel('Psi (radians')
plt.legend()
plt.show()

```

```python
#compute standard deviation of phi and psi for A and B
std_phi_A = np.std(data_A[:,1])
std_psi_A = np.std(data_A[:,2])
std_phi_B = np.std(data_B[:,1])
std_psi_B = np.std(data_B[:,2])
#print out standard deviation
print("Standard deviation of phi for A is", std_phi_A)
print("Standard deviation of psi for A is", std_psi_A)
print("Standard deviation of phi for B is", std_phi_B)
print("Standard deviation of psi for B is", std_psi_B)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mdtraj as md
```

```python
#load the grid file
data_A = np.loadtxt('../sim_A/metad.grid.dat',comments=["@","#"])
data_B = np.loadtxt('../sim_B/metad.grid.dat',comments=["@","#"])
```

```python
#plot phi vs energy
plt.plot(data_A[:,0],data_A[:,1],label="sim_A")
plt.plot(data_B[:,0],data_B[:,1],label="sim_B")
plt.xlabel("phi (radians)")
plt.ylabel("energy (kJ/mol)")
plt.title("Energy vs phi")
plt.legend()
plt.show()
```

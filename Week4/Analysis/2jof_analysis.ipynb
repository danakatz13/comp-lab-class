```python
import matplotlib.pyplot as mpl
import mdtraj as md
```

```python
traj = traj = md.load_xtc('../Setup/2jof_amber/gromacs/md_center.xtc', top='../Setup/2jof_amber/gromacs/topol_water-free.gro')
```

```python
#compute end to end distance
end_to_end = md.compute_distances(traj, [[0, traj.n_atoms-1]])

#compute radius of gyration
radius_of_gyration = md.compute_rg(traj)
```

```python
#plot end to end distance
mpl.plot(end_to_end)
mpl.xlabel('Frame')
mpl.ylabel('End to end distance (nm)')
mpl.title('End to end distance of 2jof')

```

```python
#plot radius of gyration
mpl.plot(radius_of_gyration)
mpl.xlabel('Frame')
mpl.ylabel('Radius of gyration (nm)')
mpl.title('Radius of gyration of 2jof')
```
